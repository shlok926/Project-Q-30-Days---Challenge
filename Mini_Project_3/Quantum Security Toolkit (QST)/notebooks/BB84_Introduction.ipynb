# Quantum Key Distribution: Introduction to BB84 Protocol

This notebook introduces the concepts of Quantum Key Distribution (QKD) and demonstrates the BB84 protocol implementation using the **Quantum Security Toolkit (QST)**.

## 1. What is Quantum Key Distribution (QKD)?
QKD uses the laws of quantum mechanics to secure communication channels. Unlike classical cryptography, which relies on the computational difficulty of mathematical problems, QKD's security is guaranteed by physical invariants: quantum state perturbation under measurement.

## 2. The BB84 Protocol Steps
Introduced by Charles Bennett and Gilles Brassard in 1984, the protocol works as follows:
- **State Preparation (Alice):** Alice generates random classical bits and encodes them into quantum states using randomly chosen bases: Rectilinear (Z: $|0\rangle, |1\rangle$) or Diagonal (X: $|+\rangle, |-\rangle$).
- **Transmission:** Alice sends the polarized photons to Bob through a quantum channel.
- **Measurement (Bob):** Bob measures each incoming photon in a randomly chosen basis (Z or X).
- **Basis Reconciliation (Sifting):** Alice and Bob communicate over a public classical channel to announce their basis choices. They keep bits where their bases matched, forming the **Sifted Key**.
- **Security Verification:** They compare a subset of the sifted key to calculate the **Quantum Bit Error Rate (QBER)**. If QBER is below a threshold (typically 11% to 15%), the channel is considered secure.

## 3. Running a Basic BB84 Simulation in QST

Let's execute a basic, noise-free BB84 simulation trial.

```python
import sys
from pathlib import Path

# Ensure QST package is in PYTHONPATH
sys.path.insert(0, str(Path.cwd().parent / "src"))

from qst.models.config import SimulationConfig, ProtocolType
from qst.orchestration.orchestrator import SimulationOrchestrator

# 1. Initialize configuration with deterministic seed
config = SimulationConfig(
    n_qubits=15,
    seed=42,
    interception_probability=0.0,  # Noise-free baseline
    repetitions=1,
    protocol=ProtocolType.BB84
)

# 2. Run simulation via Orchestrator
orchestrator = SimulationOrchestrator()
result = orchestrator.run_once(config)
trial = result.simulations[0]

# 3. Display key outputs
print(f"Qubits processed: {trial.n_qubits}")
print(f"Alice Bits:       {trial.alice_bits}")
print(f"Alice Bases:      {trial.alice_bases}")
print(f"Bob Bases:        {trial.bob_bases}")
print(f"Sifted Key:       {trial.sifted_key}")
print(f"Key Rate:         {trial.key_rate:.4f}")
print(f"QBER (Error):     {trial.qber:.4f}")
print(f"Security Status:  {trial.security_metrics.status.value}")
```
